```python
#loads up the meta features test set
test = pd.read_csv('meta_features_test.csv', sep=',')
test= test.drop('Unnamed: 0', 1)
print(test)
#scales and encodes relevant feature columns
std_scale= StandardScaler()

numeric_features_test = test.select_dtypes(include=[np.number])
numeric_features_test = numeric_features_test.drop('is_fraud',1)
std_scale.fit(numeric_features_test)
norm_num_test= std_scale.transform(numeric_features_test)
norm_num_test = pd.DataFrame(norm_num_test, index=numeric_features_test.index, columns=numeric_features_test.columns) 
norm_num_test.update(norm_num_test)

label_enc= LabelEncoder()

categoric_features_test = test.select_dtypes(include=['object']).copy()
categoric_cols=(categoric_features_test.columns).tolist()
categoric_features_test[categoric_cols]=categoric_features_test[categoric_cols].apply(lambda col: label_enc.fit_transform(col))
le_cat_test= categoric_features_test[categoric_cols]
#concates these to form features and target class
X_test= pd.concat([norm_num_test, le_cat_test], axis=1)
Y_test= test['is_fraud'].copy()
```

```python
import pandas as pd

datasets=['ADASYN_samples.csv','BorderlineSmote_samples.csv', 'CSMOUTE_samples.csv', 'EditedNN_samples.csv','MWMOTE_samples.csv','NCR_samples.csv','OSS_samples.csv','SMOTE_samples.csv','SMOTEENN_samples.csv']

for dataset in datasets:
#the for loop above iterates each sampled training in order. for each iteration, the train variable below loads up the dataset relevant modifications are carried out in the following lines
    train = pd.read_csv(dataset, sep=',')
    train = train.drop('Unnamed: 0', 1)

    std_scale= StandardScaler()

#extracts numeric and categoric features, scales and encodes them respectively
    numeric_features_train = train.select_dtypes(include=[np.number])
    numeric_features_train = numeric_features_train.drop('is_fraud',1)
    std_scale.fit(numeric_features_train)
    norm_num_train= std_scale.transform(numeric_features_train)
    norm_num_train = pd.DataFrame(norm_num_train, index=numeric_features_train.index, columns=numeric_features_train.columns) 
    norm_num_train.update(norm_num_train)
    print(len(norm_num_train))

    label_enc= LabelEncoder()

    categoric_features_train = train.select_dtypes(include=['object']).copy()
    print(len(categoric_features_train))
    categoric_cols=(categoric_features_train.columns).tolist()
    categoric_features_train[categoric_cols]=categoric_features_train[categoric_cols].apply(lambda col: label_enc.fit_transform(col))
    le_cat_train= categoric_features_train[categoric_cols]
    print(len(le_cat_train))

    X_train= pd.concat([norm_num_train, le_cat_train],axis=1)
    Y_train = train['is_fraud'].reset_index(drop=True)
    Y_train= pd.DataFrame(Y_train)

#Random Forest and Extra Trees used for feature selection, selects 10 best features through feature importance

    modelxgb = RandomForestClassifier()
    modelxgb.fit(X_train, Y_train)

    f_xgb = pd.DataFrame(data={'X_train':X_train.columns,'value':modelxgb.feature_importances_})
    f_xgb = f_xgb.sort_values(['value'],ascending=False )
    plt.figure(figsize=(15,8))
    sns.barplot(f_xgb['X_train'],f_xgb['value'])

    etcmodel = ExtraTreesClassifier()
    etcmodel.fit(X_train,Y_train)

    f_etc = pd.DataFrame(data={'X_train':X_train.columns,'value':etcmodel.feature_importances_})
    f_etc = f_etc.sort_values(['value'],ascending=False )
    plt.figure(figsize=(15,8))
    sns.barplot(f_etc['X_train'],f_etc['value'])

    ft = pd.merge(f_xgb, f_etc,how='inner', on=["X_train"])
    ft.sort_values(["value_x","value_y"],ascending=False, inplace=True)

    top10ft = ft.head(10)
#selects top 10 variables
    X_train=X_train[top10ft.X_train]
#we dont really need to save these variables. just did them for testing things out without the program crashing
    X_train.to_csv('X_train.csv')
    Y_train.to_csv('Y_train.csv')

    X_test=X_test[top10ft.X_train]
    X_test.to_csv('X_test.csv')
    Y_test.to_csv('Y_test.csv')
    

###########created neural network models (without running them)#########
#transfer learning models
    cnn_pt_model = load_model('cnn_model.h5') # load this model with its weights
    fcn_pt_model = load_model('fcn_model.h5') # load this model with its weights
###selects the top 3 fully connected layers from the pre-train-model.ipynb file
    top3layerModel_fcn = tf.keras.Model(
        inputs  = fcn_pt_model.inputs,
        outputs = fcn_pt_model.layers[2].output,
    )
    top3layerModel_fcn.summary()
###selects the top 6 convolutional neural layers from the pre-train-model.ipynb file
    top6layerModel_cnn = tf.keras.Model(
        inputs  = cnn_pt_model.inputs,
        outputs = cnn_pt_model.layers[5].output,
    )
    top6layerModel_cnn.summary()


    ##transfer learning models##
    #cnn transfer learning model
    transfer_model_cnn = Sequential()
    transfer_model_cnn.add(top6layerModel_cnn)
    transfer_model_cnn.add(Flatten())
    transfer_model_cnn.add(Dense(units=64, activation='relu'))
    transfer_model_cnn.add(Dropout(0.5))
    transfer_model_cnn.add(Dense(units=1, activation='sigmoid'))
    transfer_model_cnn.summary()

    #fcn transfer learning model
    transfer_model_fcn = tf.keras.Sequential()
    transfer_model_fcn.add(top3layerModel_fcn)
    transfer_model_fcn.add(tf.keras.layers.Dense(10))
    transfer_model_fcn.add(tf.keras.layers.Activation('relu'))
    transfer_model_fcn.add(tf.keras.layers.Dropout(0.2))
    transfer_model_fcn.add(tf.keras.layers.Dense(10))
    transfer_model_fcn.add(tf.keras.layers.Activation('relu'))
    transfer_model_fcn.add(tf.keras.layers.Dropout(0.2))
    transfer_model_fcn.add(tf.keras.layers.Dense(1))
    transfer_model_fcn.add(tf.keras.layers.Activation('sigmoid'))

    transfer_model_fcn.summary()

    #non-transfer learning models##
    #cnn non-transfer learning mdoels
    X_train_sim_cnn = X_train.values.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test_sim_cnn = X_test.values.reshape(X_test.shape[0], X_test.shape[1], 1)
    normal_model_cnn = Sequential()
    normal_model_cnn.add(Conv1D(filters = 32, kernel_size=2, activation='relu', input_shape=(X_train_sim_cnn[0].shape)))
    normal_model_cnn.add(BatchNormalization())
    normal_model_cnn.add(Dropout(0.2))
    normal_model_cnn.add(Conv1D(filters = 32, kernel_size=2, activation='relu'))
    normal_model_cnn.add(BatchNormalization())
    normal_model_cnn.add(Dropout(0.4))
    normal_model_cnn.add(Flatten())
    normal_model_cnn.add(Dense(units=64, activation='relu'))
    normal_model_cnn.add(Dropout(0.5))
    normal_model_cnn.add(Dense(units=1, activation='sigmoid'))
    normal_model_cnn.summary()


    #fcn non-transfer learning model
    normal_model_fcn = tf.keras.models.Sequential()
    normal_model_fcn.add(tf.keras.layers.Dense(100, input_shape=(10,)))
    normal_model_fcn.add(tf.keras.layers.Activation('relu'))
    normal_model_fcn.add(tf.keras.layers.Dropout(0.2))
    normal_model_fcn.add(tf.keras.layers.Dense(10))
    normal_model_fcn.add(tf.keras.layers.Activation('relu'))
    normal_model_fcn.add(tf.keras.layers.Dropout(0.2))
    normal_model_fcn.add(tf.keras.layers.Dense(1))
    normal_model_fcn.add(tf.keras.layers.Activation('sigmoid'))
    normal_model_fcn.summary()

    from sklearn.linear_model import LogisticRegression
    from sklearn.neighbors import KNeighborsClassifier
    from sklearn.svm import SVC
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.ensemble import ExtraTreesClassifier
    from sklearn.naive_bayes import GaussianNB
    from sklearn import model_selection
    from sklearn.utils import class_weight
    from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score
    from sklearn.model_selection import GridSearchCV

    import numpy as np
    import pandas as pd

#again, we dont need this, it was just used for testing
    X_train= pd.read_csv('X_train.csv', sep=',')
    X_train=X_train.drop('Unnamed: 0',1)

    X_test= pd.read_csv('X_test.csv', sep=',')
    X_test=X_test.drop('Unnamed: 0',1)

    Y_train= pd.read_csv('Y_train.csv', sep=',')
    Y_train=Y_train.drop('Unnamed: 0',1)

    Y_test= pd.read_csv('Y_test.csv', sep=',')
    Y_test=Y_test.drop('Unnamed: 0',1)

#loads up ml classifiers
    lgr= LogisticRegression()
    rfc=RandomForestClassifier(max_depth=2,random_state=0)
    etc= ExtraTreesClassifier()

#puts all models up in a dictionary for iteration later
    ml_models= {'logistic regression_on_': LogisticRegression(),'random_forest_on_':RandomForestClassifier(),'extra_trees_on_': ExtraTreesClassifier()} 

    dl_fcn_models= {'transfer_model_fcn_on_': transfer_model_fcn,'normal_model_fcn_on_': normal_model_fcn}

    dl_cnn_models= {'transfer_model_cnn_on_': transfer_model_cnn,'normal_model_cnn_on_': normal_model_cnn}

#used to add the financial loss metric inside the classification report csv's
    from csv import writer
    def append_list_as_row(file_name, list_of_elem):
        # Open file in append mode
        with open(file_name, 'a+', newline='') as write_obj:
            # Create a writer object from csv module
            csv_writer = writer(write_obj)
            # Add contents of list as last row in the csv file
            csv_writer.writerow(list_of_elem)

    #print('========================== ML Model Evaluation Results ========================' "\n")  
#added comments in this for loop. the comments apply to the other two for loops below this one too
    for i, v in ml_models.items():
        #fit training model
        v.fit(X_train,Y_train)
        cfr_andDataset= i+dataset
        print('model:',cfr_andDataset,' trained.')
        #use trained model to predict labels for the meta features dataset
        v_pred= v.predict(X_test)
        v_pred= pd.DataFrame(v_pred, columns=Y_test.columns)
        #saves the classification report for each model
        class_report= classification_report(Y_test, v_pred, output_dict=True)
        class_report = pd.DataFrame(class_report).transpose()
        report_to_csv= '_class_report.csv'
        class_report_csv= cfr_andDataset+report_to_csv
        class_report.to_csv(class_report_csv)
        #saves the confusion matrix for each model
        cm= confusion_matrix(Y_test, v_pred)
        cm= pd.DataFrame(cm)
        cm_csv='_cm.csv'
        cm_csv_model=cfr_andDataset+cm_csv
        cm.to_csv(cm_csv_model)
        
        #iterates the dataset to extract the false positives
        false_positives= np.logical_and(Y_test != v_pred, v_pred == 0)
        credit_fraud_data= test[false_positives['is_fraud']==True]
        #extracts the total amount lost even when using the specific classifier
        financial_loss= np.sum(credit_fraud_data['amt'])
        class_rep_fin_loss= ['finanacial loss',financial_loss]
        #appends this metric to the classification report csv file
        append_list_as_row(class_report_csv, class_rep_fin_loss)
        print("Financial loss is: ", financial_loss)
        
        #saves the labels and predictions just so that we could use this for creating more metrics
        result=pd.concat([Y_test,v_pred],1)
        result.rename(columns={'is_fraud': 'Y_test', 'is_fraud': 'is_fraud'})
        result.to_csv(cfr_andDataset)
        print('results for', cfr_andDataset, 'saved to a csv file for the specific sampling dataset')


    print('========================== DL-FCN Model Evaluation Results ========================' "\n")

    for i, v in dl_fcn_models.items():
        batch_size=32
        epochs=10
        v.compile(optimizer=Adam(lr=0.001), loss='binary_crossentropy', metrics=['accuracy'])
        v.fit(X_train, Y_train, batch_size=batch_size, epochs=epochs, verbose=1)
        cfr_andDataset= i+dataset
        print('model:',cfr_andDataset,' trained.')
        v_pred= v.predict_classes(X_test)
        v_pred= pd.DataFrame(v_pred, columns=Y_test.columns)
                
        class_report= classification_report(Y_test, v_pred, output_dict=True)
        class_report = pd.DataFrame(class_report).transpose()
        report_to_csv= '_class_report.csv'
        class_report_csv= cfr_andDataset+report_to_csv
        class_report.to_csv(class_report_csv)
        cm= confusion_matrix(Y_test, v_pred)
        cm= pd.DataFrame(cm)
        cm_csv='_cm.csv'
        cm_csv_model=cfr_andDataset+cm_csv
        cm.to_csv(cm_csv_model)
        
        
        false_positives= np.logical_and(Y_test != v_pred, v_pred == 0)
        credit_fraud_data= test[false_positives['is_fraud']==True]
        financial_loss= np.sum(credit_fraud_data['amt'])
        class_rep_fin_loss= ['finanacial loss',financial_loss]
        append_list_as_row(class_report_csv, class_rep_fin_loss)
        print("Financial loss is: ", financial_loss)

        result=pd.concat([Y_test,v_pred],1)
        result.rename(columns={'is_fraud': 'Y_test', 'is_fraud': 'is_fraud'})
        result.to_csv(cfr_andDataset)
        print('results for', cfr_andDataset, 'saved to a csv file for the specific sampling dataset')


    print('========================== DL-CNN Model Evaluation Results ========================' "\n")

    for i, v in dl_cnn_models.items():
        batch_size=32
        epochs=10 
        v.compile(optimizer=Adam(lr=0.001), loss='binary_crossentropy', metrics=['accuracy'])
        v.fit(X_train_sim_cnn, Y_train, batch_size=batch_size, epochs=epochs, verbose=1)
        cfr_andDataset= i+dataset
        print('model:',cfr_andDataset,' trained.')
        v_pred= v.predict_classes(X_test_sim_cnn)
        v_pred= pd.DataFrame(v_pred, columns=Y_test.columns)

        class_report= classification_report(Y_test, v_pred, output_dict=True)
        class_report = pd.DataFrame(class_report).transpose()
        report_to_csv= '_class_report.csv'
        class_report_csv= cfr_andDataset+report_to_csv
        class_report.to_csv(class_report_csv)
        cm= confusion_matrix(Y_test, v_pred)
        cm= pd.DataFrame(cm)
        cm_csv='_cm.csv'
        cm_csv_model=cfr_andDataset+cm_csv
        cm.to_csv(cm_csv_model)
        
        false_positives= np.logical_and(Y_test != v_pred, v_pred == 0)
        credit_fraud_data= test[false_positives['is_fraud']==True]
        financial_loss= np.sum(credit_fraud_data['amt'])
        class_rep_fin_loss= ['finanacial loss',financial_loss]
        append_list_as_row(class_report_csv, class_rep_fin_loss)
        print("Financial loss is: ", financial_loss)
        
        result=pd.concat([Y_test,v_pred],1)
        result.rename(columns={'is_fraud': 'Y_test', 'is_fraud': 'is_fraud'})
        result.to_csv(cfr_andDataset)
        print('results for', cfr_andDataset, 'saved to a csv file for the specific sampling dataset')


```
